```python
# Quantidade de serviços especializados
qtd_serv_distinto = df["SERVIÇO"].nunique()
qtd_serv_distinto
```

```python
# Criando variáveis para saber a quantidade de serviços por município.
serv_municipio = df.groupby("MUNICÍPIO")["SERVIÇO"].nunique().reset_index()
serv_por_municipio = serv_municipio.sort_values(by= "SERVIÇO", ascending=False)
```

```python
# Plotando
top_20 = serv_por_municipio.head(20)

plt.figure(figsize=(15, 6))
sns.barplot(data=top_20, x='SERVIÇO', y='MUNICÍPIO', palette='viridis')
plt.title('Top 20 Municípios Com Maior Quantidade de Serviços Especializados', fontsize=14)
plt.xlabel('Número de Serviços Únicos')
plt.ylabel('Município')
plt.tight_layout()
plt.show()
```

```python
# Criando variáveis para saber a menor quantidade de serviços por município.
serv_municipio = df.groupby("MUNICÍPIO")["SERVIÇO"].nunique().reset_index()
serv_por_municipio_menor = serv_municipio.sort_values(by= "SERVIÇO", ascending=True)

top_20 = serv_por_municipio_menor.head(20)

plt.figure(figsize=(15, 6))
sns.barplot(data=top_20, x='SERVIÇO', y='MUNICÍPIO', palette='viridis')
plt.title('Os 20 Municípios Com Menor Quantidade de Serviços Especializados', fontsize=14)
plt.xlabel('Número de Serviços Únicos')
plt.ylabel('Município')
plt.tight_layout()
plt.show()
```

```python
# Percentual dos estabelecimentos por tipo de natureza jurídica
estab_publicos = df[df["CATEGORIA NATUREZA JURÍDICA"] == "PÚBLICO"]["NOME FANTASIA"].nunique()
estab_privados = df[df["CATEGORIA NATUREZA JURÍDICA"] == "PRIVADO"]["NOME FANTASIA"].nunique()
estab_slucro = df[df["CATEGORIA NATUREZA JURÍDICA"] == "SEM FINS LUCRATIVOS"]["NOME FANTASIA"].nunique()

plt.figure(figsize=(6, 6))
plt.pie([estab_publicos, estab_privados, estab_slucro],
        labels=['Públicos', 'Privados', 'Sem fins lucrativos'],
        colors=['#3498db', '#e74c3c', '#2ecc71'],
        autopct='%1.1f%%',
        startangle=90)
plt.title('Proporção de Estabelecimentos Públicos e Privados')
plt.show()
```

```python
# Identificação de hospitais e ambulatórios na base de dados.
qtd_ambulatorios = df[df["TIPO NOVO DO ESTABELECIMENTO"].str.contains("AMBULATÓRIO|AMBULATORIO", case=False, na=False)].shape[0]
qtd_hospitais = df[df["TIPO NOVO DO ESTABELECIMENTO"].str.contains("HOSPITAL", case=False, na=False)].shape[0]


print(f"Ambulatórios: {qtd_ambulatorios}")
print(f"Hospitais: {qtd_hospitais}")
print(f"Total: {qtd_ambulatorios + qtd_hospitais}")
```

```python
# Plotando
plt.figure(figsize=(8, 8))
plt.pie([qtd_ambulatorios, qtd_hospitais],
        labels=['Ambulatórios', 'Hospitais'],
        colors=['#3498db', '#e74c3c'],
        autopct='%1.1f%%',
        startangle=90)
plt.title('Proporção de Ambulatórios e Hospitais')
plt.show()
```

```python
#Criando variáveis para hospiatais e ambulatórios por município.
ambulatorios = df[df["TIPO NOVO DO ESTABELECIMENTO"].str.contains("AMBULATÓRIO|AMBULATORIO", case=False, na=False)]["MUNICÍPIO"].value_counts()
hospitais= df[df["TIPO NOVO DO ESTABELECIMENTO"].str.contains("HOSPITAL", case=False, na=False)]["MUNICÍPIO"].value_counts()

dados = pd.DataFrame({'Ambulatórios': ambulatorios,'Hospitais': hospitais}).fillna(0).astype(int)
```

```python
top_municipios = dados.sum(axis=1).nlargest(20).index
dados_top = dados.loc[top_municipios]

plt.figure(figsize=(14, 8))
bar_width = 0.35
x = np.arange(len(dados_top))

# 5. Criar as barras
bars_amb = plt.bar(x - bar_width/2, dados_top['Ambulatórios'], bar_width,
                  color='#3498db', label='Ambulatórios', edgecolor='black')

bars_hosp = plt.bar(x + bar_width/2, dados_top['Hospitais'], bar_width,
                   color='#e74c3c', label='Hospitais', edgecolor='black')

# 6. Personalização
plt.title('Distribuição de Ambulatórios e Hospitais por Município (Top 15)',
         fontsize=16, pad=20)
plt.xlabel('Município', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.xticks(x, dados_top.index, rotation=45, ha='right')
plt.legend()

# 7. Adicionar valores nas barras
for bars in [bars_amb, bars_hosp]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height,
                f'{int(height)}',
                ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
